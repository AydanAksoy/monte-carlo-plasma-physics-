```python
import matplotlib.pyplot as plt
import numpy as np
import random
import math

# симулируем 1000 частиц / simulate 1000 particles
results = []
for particle in range(1000):
    position = 0
    for step in range(20):
        coin = random.choice([-1, +1])
        position = position + coin
    results.append(position)

# рисуем график / draw plot
plt.figure(figsize=(10, 6))
plt.hist(results, bins=20, color='steelblue', 
         edgecolor='white', alpha=0.8)
plt.xlabel('позиция / position')
plt.ylabel('количество частиц / number of particles')
plt.title('Монте-Карло: 1000 ионов / Monte Carlo: 1000 ions')
plt.axvline(x=0, color='red', linestyle='--', 
            label='центр / centre')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

# симулируем / simulate
results = []
lost = []
safe = []

for particle in range(1000):
    position = 0
    for step in range(20):
        coin = random.choice([-1, +1])
        position = position + coin
    results.append(position)
    if position > 5 or position < -5:
        lost.append(position)
    else:
        safe.append(position)

# график / plot
plt.figure(figsize=(10, 6))
plt.hist(safe, bins=20, color='steelblue', 
         alpha=0.8, label='безопасно / safe')
plt.hist(lost, bins=20, color='red', 
         alpha=0.8, label='потеряно / lost')
plt.axvline(x=5, color='black', linewidth=2,
            linestyle='--', label='стенка / wall')
plt.axvline(x=-5, color='black', linewidth=2,
            linestyle='--')
plt.xlabel('позиция / position')
plt.ylabel('количество / count')
plt.title('Потери ионов / Ion losses')
plt.legend()
plt.show()

print("потеряно / lost:", len(lost), "из / of 1000")
print("процент / percent:", round(len(lost)/10, 1), "%")
```
