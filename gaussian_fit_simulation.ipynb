```python
import matplotlib.pyplot as plt
import numpy as np

# симуляция / simulation
positions = np.random.choice([-1, +1], size=(1000, 20))
final_positions = np.sum(positions, axis=1)

mean = np.mean(final_positions)
sigma = np.std(final_positions)

# график / plot
fig, ax = plt.subplots(figsize=(10, 6))

# столбики / bars
ax.hist(final_positions, bins=25, density=True,
        color='steelblue', edgecolor='white', 
        alpha=0.7, label='симуляция / simulation')

# кривая Гаусса / Gaussian curve
x = np.linspace(-20, 20, 300)
gaussian = (1/(sigma * np.sqrt(2*np.pi))) * \
            np.exp(-0.5*((x-mean)/sigma)**2)
ax.plot(x, gaussian, 'red', linewidth=2.5,
        label=f'Гаусс / Gaussian (σ={sigma:.2f})')

ax.axvline(x=mean, color='green', 
           linestyle='--', linewidth=1.5,
           label=f'среднее / mean = {mean:.2f}')

ax.set_xlabel('позиция / position')
ax.set_ylabel('плотность / density')
ax.set_title('Монте-Карло + Гаусс / Monte Carlo + Gaussian')
ax.legend()
plt.show()
```
