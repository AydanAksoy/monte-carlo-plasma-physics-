```python
import numpy as np

# numpy делает случайные числа быстро
# numpy makes random numbers fast

positions = np.random.choice([-1, +1], size=(1000, 20))
final_positions = np.sum(positions, axis=1)

print("первые 10 результатов / first 10 results:")
print(final_positions[:10])
print()
print("среднее / average:", round(np.mean(final_positions), 2))
print("сигма / sigma:", round(np.std(final_positions), 2))
```

```python
import matplotlib.pyplot as plt
import numpy as np

# симуляция / simulation
positions = np.random.choice([-1, +1], size=(1000, 20))
final_positions = np.sum(positions, axis=1)

# график / plot
plt.figure(figsize=(10, 6))
plt.hist(final_positions, bins=25, 
         color='steelblue', edgecolor='white', alpha=0.8)
plt.xlabel('позиция / position')
plt.ylabel('количество / count')
plt.title(f'numpy симуляция / numpy simulation\n'
          f'среднее={np.mean(final_positions):.2f}, '
          f'σ={np.std(final_positions):.2f}')
plt.axvline(x=0, color='red', linestyle='--')
plt.show()
```
